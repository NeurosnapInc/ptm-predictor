```python
import pandas as pd
import glob
import os

# Get all CSV files in the modified_residues folder
csv_files = glob.glob('./crosslinks/*.csv')

for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    df_filtered = df[df['sequence'].str.len() <= 1022]
    df_processed = df_filtered.drop_duplicates()
    base_name = os.path.splitext(os.path.basename(csv_file))[0]
    output_file = f'./crosslinks/{base_name}_processed.csv'
    df_processed.to_csv(output_file, index=False)
    print(f"Processed {os.path.basename(csv_file)}: {len(df)} -> {len(df_processed)} rows")
```

```python
import pandas as pd
import glob
import re

def check_position_patterns(folder_path, column_name='position'):
    pattern = r'^(\d+)(\.\.(\d+))?$'
    csv_files = glob.glob(f'{folder_path}/*_processed.csv')
    for csv_file in csv_files:
        df = pd.read_csv(csv_file)
        
        print(f"\nChecking {csv_file}:")
        print("-" * 50)
        
        found_non_matching = False
        contains_double_dots = 0
        single_numbers = 0
        
        for idx, value in df[column_name].items():
            value = str(value)
            if not re.match(pattern, value):
                print(f"Row {idx}: '{value}' does not match pattern")
                found_non_matching = True
            else:
                if '..' in value:
                    contains_double_dots += 1
                else:
                    single_numbers += 1
        
        if not found_non_matching:
            print("All values match the expected pattern!")
        
        print(f"\nSummary for {csv_file}:")
        print(f"  Double numbers: {contains_double_dots}")
        print(f"  Single numbers: {single_numbers}")
```

```python
import pandas as pd
import glob
import os

def process_csv_files(csv_files):
    for csv_file in csv_files:
        df = pd.read_csv(csv_file)
        print(len(df))
        expanded_rows = []
        for index, row in df.iterrows():
            position = str(row['position'])
            if '..' in position:
                parts = position.split('..')
                for part in parts:
                    if ':' in part:
                        processed_part = part.split(':')[-1]
                    else:
                        processed_part = part
                    if processed_part.strip().isdigit():
                        cur_row = row.copy()
                        cur_row['position'] = int(processed_part)
                        expanded_rows.append(cur_row)
            else:
                if ':' in position:
                    processed_part = position.split(':')[-1]
                else:
                    processed_part = position
                if processed_part.strip().isdigit():
                    cur_row = row.copy()
                    cur_row['position'] = int(processed_part)
                    expanded_rows.append(cur_row)
        new_df = pd.DataFrame(expanded_rows)
        print(len(new_df))
        directory = os.path.dirname(csv_file)
        filename = os.path.basename(csv_file)
        prefix = filename.split('_processed')[0]
        output_file = os.path.join(directory, prefix + '_split.csv')
        new_df.to_csv(output_file, index=False)
        print(f"Saved: {output_file}")
```

```python
import pandas as pd
import glob
import os

folders = ['crosslinks', 'disulfide_bonds', 'glycosylations', 'lipidations', 'modified_residues']

for folder in folders:
    csv_files = glob.glob(f'./{folder}/*_split.csv')
    
    for csv_file in csv_files:
        df = pd.read_csv(csv_file)
        
        # Drop duplicates
        df_final = df.drop_duplicates()
        
        # Create output filename
        directory = os.path.dirname(csv_file)
        filename = os.path.basename(csv_file)
        base_name = filename.replace('_split.csv', '')
        output_file = os.path.join(directory, base_name + '_final.csv')
        
        # Save the deduplicated dataframe
        df_final.to_csv(output_file, index=False)
```

```python
import pandas as pd
import glob

# Define the folders to search
folders = ['crosslinks', 'disulfide_bonds', 'glycosylations', 'lipidations', 'modified_residues']

# Loop through each folder
for folder in folders:
    # Get all _final.csv files in the current folder
    csv_files = glob.glob(f'./{folder}/*_final.csv')
    
    for csv_file in csv_files:
        df = pd.read_csv(csv_file)
        if 'position' in df.columns and 'sequence_len' in df.columns:
            count = (df['position'] > df['sequence_len']).sum()
            print(f"{csv_file}: {count} rows where position > sequence_len")
        else:
            print(f"{csv_file}: missing 'position' or 'sequence_len' column")
```

```python
import pandas as pd
import glob

# Define the folders to search
folders = ['crosslinks', 'disulfide_bonds', 'glycosylations', 'lipidations', 'modified_residues']

# Loop through each folder
for folder in folders:
    # Get all _final.csv files in the current folder
    csv_files = glob.glob(f'./{folder}/*_final.csv')
    
    for csv_file in csv_files:
        df = pd.read_csv(csv_file)
        df_cleaned = df[df['position'] <= df['sequence_len']]
        output_path = csv_file.replace('_final.csv', '_final1.csv')
        df_cleaned.to_csv(output_path, index=False)
        print(f"Saved cleaned file: {output_path} ({len(df) - len(df_cleaned)} rows removed)")
```

```python
import pandas as pd
import glob

folders = ['crosslinks', 'disulfide_bonds', 'glycosylations', 'lipidations', 'modified_residues']
for folder in folders:
    csv_files = glob.glob(f'./{folder}/*_final1.csv')
    for csv_file in csv_files:
        df = pd.read_csv(csv_file)
        position_rows = []
        for seq, group in df.groupby('sequence'):
            positions = []
            for pos in group['position']:
                positions.append(int(pos)-1)
            positions = list(set(positions))
            positions.sort()
            positions_str = ','.join(map(str, positions))
            position_rows.append({
                'sequence': seq, 
                'positions': positions_str, 
                'num_positions': len(positions)
            })
        # Create a DataFrame with the results
        positions_df = pd.DataFrame(position_rows)
        # Save to file
        output_path = csv_file.replace('_final1.csv', '_positions.csv')
        positions_df.to_csv(output_path, index=False)
        print(f"Saved positions file: {output_path}")
```

```python
import pandas as pd
import glob
import os
import shutil

folders = ['crosslinks', 'disulfide_bonds', 'glycosylations', 'lipidations', 'modified_residues']
for folder in folders:
    dest_folder = os.path.join('data_final', folder)
    
    csv_files = glob.glob(f'./{folder}/*_positions.csv')
    for csv_file in csv_files:
        filename = os.path.basename(csv_file)
        dest_path = os.path.join(dest_folder, filename)
        shutil.move(csv_file, dest_path)
        print(f"Moved {csv_file} to {dest_path}")
```

```python
import pandas as pd
import glob

folders = ['crosslinks', 'disulfide_bonds', 'glycosylations', 'lipidations', 'modified_residues']
csv_totals = []

for folder in folders:
    csv_files = glob.glob(f'./data_final/{folder}/*.csv')
    
    for csv_file in csv_files:
        df = pd.read_csv(csv_file)
        total_positions = df['num_positions'].sum()
        csv_totals.append((csv_file, total_positions))

# Sort by total positions in descending order
csv_totals.sort(key=lambda x: x[1], reverse=True)

print("CSV files sorted by total positions (descending):")
for csv_file, total in csv_totals:
    print(f"{csv_file}: {total}")
```
